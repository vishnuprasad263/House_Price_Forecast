```python
#MODEL BUILDING
from sklearn.linear_model import LinearRegression

reg=LinearRegression()
labels = data1['price']
conv_dates = [1 if values == 2014 else 0 for values in data1.date ]
data1['date'] = conv_dates
train1 = data1.drop(['id', 'price'],axis=1)
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(train1, labels, test_size=0.1, random_state=2)
```

```python
reg.fit(x_train,y_train)
```

```python
reg.score(x_test,y_test)
```

```python
from sklearn import ensemble
clf = ensemble.GradientBoostingRegressor(n_estimators = 400, max_depth = 5, min_samples_split = 2,
          learning_rate = 0.1, loss = 'squared_error')
```

```python
clf.fit(x_train, y_train)
```

```python
clf.score(x_test,y_test)
```

```python
params = clf.get_params()
t_sc = np.zeros((params['n_estimators']),dtype=np.float64)
```

```python
y_pred = reg.predict(x_test)
```

```python
for i,y_pred in enumerate(clf.staged_predict(x_test)):
    t_sc[i]=clf.loss_(y_test,y_pred)
```

```python
testsc = np.arange((params['n_estimators']))+1
```

```python
plt.figure(figsize=(12, 6))
plt.subplot(1, 2, 1)
plt.plot(testsc,clf.train_score_,'b-',label= 'Set dev train')
plt.plot(testsc,t_sc,'r-',label = 'set dev test')
```
